<h1>The most appropriate location to open a restaurant in New Delhi (India)<\h1><br>

<h4> Loading Postal Code Data of New Delhi <br>

```python
import pandas as pd
df=pd.read_csv("ALLINDIA.csv") #Data downloaded from Open Government Data platform India
df.head()
```

```python
df=df[df["statename"]=="DELHI"]
df=df[["pincode","Districtname","officename"]]
df.head()
```

```python
df.reset_index(inplace=True)
df.head()
```

```python
df.drop(["index"],axis=1,inplace=True)
df.head()
```

```python
print("Shape of dataframe: ",df.shape)
```

<h4>Cleaning dataframe to get all the locality under one Pincode<br><br>

```python
df.rename(columns={"pincode":'PostalCode',"Districtname":'Region',"officename":'Localities'},inplace=True)
df = df.groupby(['PostalCode', 'Region'])['Localities'].apply(lambda x:','.join(x)).reset_index()
df.drop_duplicates("PostalCode",inplace=True)
df.reset_index(inplace=True)
df.drop(["index"],axis=1,inplace=True)
df
```

```python
print("Shape of dataframe: ",df.shape)
```

```python
df=pd.concat([df, loc[["Neighborhood","Latitude","Longitude"]]], axis=1)
df.head(10)
```
